## Exploratory Data Analysis
### Objective of this Analysis:

To understand how the student's performance (test scores) is affected by the other variables (Gender, Ethnicity, Parental level of education, Lunch, Test preparation course).


```python
import pandas as pd
sp=pd.read_csv('StudentsPerformance.csv')
sp.head()
```

```python
print(sp.shape)
```

the given data is a sample of 1000 students with 5 criterias on 3 scores

```python
sp.isnull().sum()
```

no NAN values

```python
sp.describe(include='all')
```

###### Making changes in the dataframe

```python

import numpy as np

sp.rename(columns={'race/ethnicity':'race','parental level of education':'parent_edu','test preparation course':'test_prep','math score':'math_score','reading score':'reading_score','writing score':'writing_score'},inplace=True)

s = ['4-36','37-69','70-102']

sp['math_range'] = pd.cut(sp.math_score,range(4,104,33),right=False, labels=s)

sp['reading_range'] = pd.cut(sp.reading_score,range(4,104,33),right=False, labels=s)

sp['writing_range'] = pd.cut(sp.writing_score,range(4,104,33),right=False, labels=s)

sp['total']=sp.math_score + sp.reading_score + sp.writing_score

sp['total_range'] = pd.cut(sp.math_score,range(4,104,33),right=False, labels=s)


sp.head()
```

```python
sp.dtypes
```

###### The variables which act as factors affecting a student's score (gender,race,parent's educatio,lunch and test preparation) are categorical variables.
###### The scores are numerical or quantitative variables.
###### The marks range are categorical variables created for the purpose of analysis of data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Gender

```python
no_g=sp.gender.nunique()
value_g=sp.gender.unique()
count_g=sp.gender.value_counts(normalize=True)
print('The',no_g,'unique values under gender are',value_g,'with a percentage \n',count_g*100)
```

```python
sp.gender.value_counts().plot(kind='bar',align='center',width=0.2,alpha=0.5,color='blue')
plt.title('Gender count graph')
plt.ylabel('count')
plt.ylim(480,520)
print('Female count is more than Male by:',51.8-48.2,'% in the given sample of data')
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = plt.subplot2grid((2,4), (0,0))
ax2 = plt.subplot2grid((2,4), (0,1))

plt.title('$Total\ score\ data$')
sns.boxplot(x='gender',y='total',data=sp,color='green' ,ax=ax2)

sns.countplot(x='total_range',data=sp,hue='gender',ax=ax1)

print('Females have scored better than Males')
print('However greater number of female students have failed ')
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = plt.subplot2grid((2,4), (0,0))
ax2 = plt.subplot2grid((2,4), (0,1))
sns.countplot(x='math_range',data=sp,hue='gender',ax=ax1)
sns.boxplot(x='gender',y='math_score',data=sp,color='lightblue',ax=ax2)
plt.title('$GENDER\ Vs\ MATHEMATICS\ SCORE$')
print('Male score higher than Female in mathematics')
print('More number of females have failed in mathematics')
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = plt.subplot2grid((2,4), (0,0))
ax2 = plt.subplot2grid((2,4), (0,1))
sns.countplot(x='reading_range',data=sp,hue='gender',ax=ax1)
sns.boxplot(x='gender',y='reading_score',data=sp,color='aqua',ax=ax2)
plt.title('$GENDER\ Vs\ READING\ SCORE$')
print('Female score higher than Male in reading')
print('more males have failed in reading')
```

```python
fig = plt.figure(figsize=(20, 10))
ax1 = plt.subplot2grid((2,4), (0,0))
ax2 = plt.subplot2grid((2,4), (0,1))
sns.countplot(x='writing_range',data=sp,hue='gender',ax=ax1)
sns.boxplot(x='gender',y='writing_score',data=sp,color='turquoise',ax=ax2)
plt.title('$GENDER\ Vs\ WRITING\ SCORE$')
print('Female score higher than Male in writing')
print('more males have failed in writing')
```

#### Female students have on an average scored better than male students.
#### Male students lag behind Female students in both reading and writing scores. 
#### Female students lag behind Male students in Mathematics.
#### More number of female students have failed in total score.
